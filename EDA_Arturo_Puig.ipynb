```python
# Primer dataset
#df.head(2)
%matplotlib inline
import seaborn as sns

#print(df.groupby("edad_r")["p5"].value_counts())
#df = df.rename(columns={"p5": "Libros leídos en un año"})
df[df["edad_r"] == "18-24"]["p5"].value_counts()
conteos = df.groupby("edad_r")["p5"].count()
#print(columna_total)
#print(columna_total.sum())
print(round((conteos / conteos.sum()) * 100, 0))
#df.loc[0:200, ["edad_r", "p5"]]
#df.columns


edad="18-24"
conteos=df[df["edad_r"] == edad]["p5"].value_counts()
conteos_r=round((conteos / conteos.sum()) * 100, 0)
print(conteos_r)


def grafico (conteos_r,edad):
    conteos_r.plot(kind="bar", color="skyblue", figsize=(38,10))
    plt.title("Libros leídos en un año por grupo de edad")
    plt.xlabel(f"Rango de edad={edad}")
    plt.ylabel("Porcentaje de respuestas")
    plt.xticks(rotation=45)
    plt.grid(axis="y", linestyle="--", alpha=0.6)
    plt.show()

grafico(conteos_r,edad)


```

```python
edad="25-34"
conteos=df[df["edad_r"] == edad]["p5"].value_counts()
conteos_r=round((conteos / conteos.sum()) * 100, 0)
grafico(conteos_r,edad)
```
